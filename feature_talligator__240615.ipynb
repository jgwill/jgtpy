```python


from jgtpy import JGTIDS as ids,JGTIDSRequest as RQ
from jgtpy import jgtapyhelper as tah
from jgtpy import JGTPDSP as pds


i="SPX500";t="D1"

quote_count=-1

talligator_flag=True
balligator_flag=True
# if talligator_flag:
#   quote_count=1200
```

```python

  
# dfsrc = pds.getPH(i,t,use_fresh=False,quote_count=quote_count)
use_fresh = False
use_full=False
refresh_data=False

if refresh_data:
  dfsrc = pds.getPH(i,t,use_fresh=use_fresh,talligator_flag=talligator_flag,use_full=use_full)
else:
  povfn=pds.get_pov_local_data_filename(i,t,use_full=use_full)
  dfsrc=pds.read_ohlc_df_from_file(povfn)
```

```python

dfsrc2=dfsrc.copy()
dfsrc
```

```python


rq = RQ.JGTIDSRequest(i,t)
rq.talligator_flag=talligator_flag
rq.balligator_flag=balligator_flag
```

```python

res_tah= tah.toids(dfsrc,rq=rq)
```

```python


res_tah.columns
```

```python


use_IDS_v2_from_jgtapyhelper = True
res_jgtids = ids.tocds(dfsrc2,rq=rq,use_v2=use_IDS_v2_from_jgtapyhelper)

```
